```python
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Setup the directory, model names, and sampling names
data_dir = "/Users/samskanderi/Documents/Research_Project/SWANSF/code/results/"
all_model_names = ['KNN', 'SVM', 'RandomForest', 'MLPClassifier', 'RNN', '1DCNN', 'LSTM', 'GRU']
sampling_names = ['Smote', 'Adasyn', 'GNI', 'TimeGAN', 'RusTomekSmote', 'RusTomekAdasyn', 'RusTomekGNI', 'RusTomekTimeGAN']

# Function to generate file names based on the model and sampling technique
def get_file_name(model, sampling):
    if model in ['KNN', 'SVM', 'RandomForest', 'MLPClassifier']:
        if sampling == 'Without Sampling':
            return f"{model}_NewFeatures_Results.pkl"
        else:
            return f"{model}_{sampling}_WithoutC_NewFeatures_Results.pkl"
    else:
        if sampling == 'Without Sampling':
            return f"{model}_Raw_Concatenation_Results.pkl"
        else:
            return f"{model}_{sampling}_WithoutC_Concatenation_Results.pkl"

# Prepare lists to collect all TSS scores and corresponding model labels
tss_scores = []
models = []
sampling_techniques = []

# Loop over each model and sampling technique, load the data
for model in all_model_names:
    for sampling in sampling_names:
        file_name = get_file_name(model, sampling)
        file_path = data_dir + file_name
        try:
            with open(file_path, 'rb') as file:
                data = pickle.load(file)
                tss_scores.extend([entry[6] for entry in data])  # Extracting TSS scores
                models.extend([model for _ in data])  # Repeating model name for each score
                if sampling == 'Smote':
                    sampling_techniques.extend(['SMOTE' for _ in data])  # Repeating sampling technique for each score
                elif sampling == 'Adasyn':
                    sampling_techniques.extend(['ADASYN' for _ in data])
                elif sampling == 'RusTomekSmote':
                    sampling_techniques.extend(['RUS-TL-SMOTE' for _ in data])
                elif sampling == 'RusTomekAdasyn':
                    sampling_techniques.extend(['RUS-TL-ADASYN' for _ in data])
                elif sampling == 'RusTomekGNI':
                    sampling_techniques.extend(['RUS-TL-GNI' for _ in data])
                elif sampling == 'RusTomekTimeGAN':
                    sampling_techniques.extend(['RUS-TL-TimeGAN' for _ in data])
                else:
                    sampling_techniques.extend([sampling for _ in data])  # Repeating sampling technique for each score
        except FileNotFoundError as e:
            print(f"Error opening file {file_path}: {e}")

# Create DataFrame
df = pd.DataFrame({'Model': models, 'TSS': tss_scores, 'Sampling Technique': sampling_techniques})

# Categorize sampling techniques
df['Sampling Category'] = df['Sampling Technique'].apply(lambda x: 'Over-sampling' if x in ['SMOTE', 'ADASYN', 'GNI', 'TimeGAN'] else 'Over- & Under-sampling')

# Plotting
sns.set_style("darkgrid")
plt.figure(figsize=(12, 2.5))
sns.boxplot(y='Sampling Category', x='TSS', data=df, width=0.4)
plt.title('TSS Scores: Over-sampling vs. Over- & Under-sampling', fontsize=14)
plt.ylabel('Sampling Category', fontsize=12)
plt.xlabel('TSS Score', fontsize=12)
plt.yticks(fontsize=12)
plt.xticks(fontsize=11)

plt.tight_layout()
plt.savefig('ovsou_result.pdf')

plt.show()
```

```python
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Setup directory, model names, and sampling names
data_dir = "/Users/samskanderi/Documents/Research_Project/SWANSF/code/results/"
all_model_names = ['KNN', 'SVM', 'RandomForest', 'MLPClassifier', 'RNN', '1DCNN', 'LSTM', 'GRU']
sampling_names = ['Smote', 'Adasyn', 'GNI', 'TimeGAN', 'RusTomekSmote', 'RusTomekAdasyn', 'RusTomekGNI', 'RusTomekTimeGAN']
train_test_combinations = ['P1-P2', 'P2-P3', 'P3-P4', 'P4-P5']

# Function to generate file names based on the model and sampling technique
def get_file_name(model, sampling):
    if model in ['KNN', 'SVM', 'RandomForest', 'MLPClassifier']:
        if sampling == 'Without Sampling':
            return f"{model}_NewFeatures_Results.pkl"
        else:
            return f"{model}_{sampling}_WithoutC_NewFeatures_Results.pkl"
    else:
        if sampling == 'Without Sampling':
            return f"{model}_Raw_Concatenation_Results.pkl"
        else:
            return f"{model}_{sampling}_WithoutC_Concatenation_Results.pkl"

# Lists to collect data
tss_scores = []
models = []
sampling_techniques = []
train_test_combos = []

# Load the data
for model in all_model_names:
    for sampling in sampling_names:
        file_name = get_file_name(model, sampling)
        file_path = data_dir + file_name
        try:
            with open(file_path, 'rb') as file:
                data = pickle.load(file)
                for i, entry in enumerate(data):
                    tss_scores.append(entry[6])  # Extracting TSS scores
                    models.append(model)  # Repeating model name for each score
                    sampling_techniques.append(sampling)  # Repeating sampling technique for each score
                    # Determine the train-test combination based on index
                    train_test_combos.append(train_test_combinations[i % 4])
        except FileNotFoundError as e:
            print(f"Error opening file {file_path}: {e}")

# Create DataFrame
df = pd.DataFrame({
    'Model': models,
    'TSS': tss_scores,
    'Sampling Technique': sampling_techniques,
    'Train-Test': train_test_combos
})

# Plotting
sns.set_style("darkgrid")
plt.figure(figsize=(16, 10))
sns.boxplot(data=df, y='TSS', x='Sampling Technique', hue='Train-Test', width=0.6, dodge=True)
plt.title('TSS Scores Across Train-Test Combinations and Sampling Techniques', fontsize=20)
plt.xlabel('Sampling Technique', fontsize=16)
plt.ylabel('TSS Score', fontsize=16)
plt.xticks(rotation=45)
plt.legend(title='Train-Test', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
